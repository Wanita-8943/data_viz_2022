##ตารางข้อมูล covid ปี2022

```python
data_covid2022 =  pd.read_csv(os.path.join(path,'confirmed-cases-since-280265.csv'))
data_covid2022
```

##ต่อตาราง

```python
covid = pd.concat([data_covid2020,data_covid2021_1,data_covid2021_2,data_covid2022])
covid
```

###เลือกเอา column ที่ต้องการ

```python
Covid = covid[['No.','announce_date','sex','age','Unit','province_of_isolation']]
Covid
```

###ตรวจสอบค่า Missing

```python
Covid.isnull().any()
```

###แทนค่า mising

```python
Covid[['sex']] = Covid[['sex']].fillna('ไม่ระบุ')
Covid[['Unit']] = Covid[['Unit']].fillna('ปี')
Covid[['province_of_isolation']] = Covid[['province_of_isolation']].fillna('ไม่ระบุ')
```

```python
Covid = Covid.fillna(method="bfill") #ใช้ bfill คือการแทนที่ด้วยค่าถัดไปคือการนำค่าถัดไปที่มีค่า ของเซลล์ตัวเองมาแทนที่ลงในช่องที่เป็น NaN
Covid.isnull().any() #เช็คข้อมูลในแต่ละคอลัมน์อีกครั้ง
```

```python
Covid
```

```python
set(Covid['province_of_isolation'])
```

```python
#ทำข้อมูลให้เหมือนกัน
Covid['province_of_isolation'].replace({'กทม':'กรุงเทพมหานคร','ขอนแก่น\xa0':'ขอนแก่น', 'ภููเก็ต':'ภูเก็ต'}, inplace=True) 
```

```python
len(set(Covid['province_of_isolation']))
```

```python
Covid_gb = Covid.groupby('province_of_isolation').size().reset_index(name='จำนวนผู้รักษาตัว')  #สร้างตารางที่รวมจำนวนผู้รักษาตัวในแต่ละจังหวัด
Covid_gb
```

```python
datacovid_dropped = Covid.dropna() #drop ค่าว่างออกจากตาราง
datacovid_dropped
```

##รวมข้อมูล Covid กับ จำนวนผู้รักษาตัวของแต่ละจังหวัด

```python
covid_iso = datacovid_dropped.merge(Covid_gb,how='left',left_on='province_of_isolation',right_on = 'province_of_isolation') 
covid_iso
```
