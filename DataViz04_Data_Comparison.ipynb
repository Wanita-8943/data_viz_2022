##1. plot bar chart เปรียบเทียบ Sepal Length ,Sepal Width ,Petal Length ,Petal Width


```python
plt.figure(figsize=[10,6]) #กำหนดขนาดของรูปในแนวตั้งและแนวนอน
width = 0.2

plt.bar([1-width,2-width,3-width,4-width],[sample_data['SepalLength'][0],sample_data['SepalWidth'][0],sample_data['PetalLength'][0],sample_data['PetalWidth'][0]],width,color='#8DC6BF',label='setosa') 
plt.bar([1,2,3,4],[sample_data['SepalLength'][1],sample_data['SepalWidth'][1],sample_data['PetalLength'][1],sample_data['PetalWidth'][1]],width,color='#FCBC66',label='versicolor') #ดอกที่ 2 versicolor	
plt.bar([1+width,2+width,3+width,4+width],[sample_data['SepalLength'][2],sample_data['SepalWidth'][2],sample_data['PetalLength'][2],sample_data['PetalWidth'][2]],width,color='#F97B4F',label='virginica') 

plt.title('plot bar chart เปรียบเทียบ Sepal Length ,Sepal Width ,Petal Length ,Petal Width');
plt.ylabel('Size (cm)')
plt.xticks([1,2,3,4],['SepalLength','SepalLength','PetalLength','PetalWidth'])
plt.legend();
```

##2. plot bar chart เพื่อเปรียบเทียบขนาดโดยรวมของดอก Iris 3 ชนิด โดยดูจาก Sepal Length ,Sepal Width ,Petal Length และ Petal Width

```python
plt.figure(figsize=[10,6]) #กำหนดขนาดของรูปในแนวตั้งและแนวนอน
width = 0.2

plt.bar([1,2,3],sample_data['SepalLength'],color='#4097AA',label='Sepal Length')
plt.bar([1,2,3],sample_data['SepalWidth'],bottom = sample_data['SepalLength'],color='#8DC6BF',label='Sepal Width')
plt.bar([1,2,3],sample_data['PetalLength'],bottom = sample_data['SepalLength']+ sample_data['SepalWidth'],color='#FCBC66',label='Petal Length') #ดอกที่ 2 versicolor	
plt.bar([1,2,3],sample_data['PetalWidth'],bottom = sample_data['PetalLength'] + sample_data['SepalLength'] + sample_data['SepalWidth'],color='#F97B4F',label='Petal Width')

plt.title('plot bar chart เพื่อเปรียบเทียบขนาดโดยรวมของดอก Iris 3 ชนิด');
plt.ylabel('Size (cm)')
plt.xticks([1,2,3],sample_data.reset_index()['Name'])
plt.legend();
```
