```python
data = data_ori[['sex','age','risk']] #à¹€à¸¥à¸·à¸­à¸à¹€à¸­à¸²à¹€à¸‰à¸žà¸²à¸°à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸—à¸µà¹ˆà¸•à¹‰à¸­à¸‡à¸à¸²à¸£à¹ƒà¸Šà¹‰
data
```

```python
data.isnull().any() #à¹€à¸Šà¸„à¸§à¹ˆà¸²à¸ à¸²à¸¢à¹ƒà¸™à¸•à¸²à¸£à¸²à¸‡à¸¡à¸µà¸„à¹ˆà¸²à¸§à¹ˆà¸²à¸‡à¸«à¸£à¸·à¸­à¹„à¸¡à¹ˆ
```

```python
data_risk1 = data.dropna() #à¹ƒà¸Šà¹‰ dropna à¹€à¸žà¸·à¹ˆà¸­à¸•à¹‰à¸­à¸‡à¸à¸²à¸£à¸¥à¸šà¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸—à¸±à¹‰à¸‡à¹à¸–à¸§à¸­à¸­à¸à¹€à¸¡à¸·à¹ˆà¸­à¸¡à¸µà¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œà¹ƒà¸”à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œà¸«à¸™à¸¶à¹ˆà¸‡à¹€à¸›à¹‡à¸™ NaN
data_risk1
```

```python
data_risk1.isnull().any()
```

```python
removes = data.shape[0] - data_risk1.shape[0]
print(f'size before drop = {data.shape[0]}')
print(f'size after drop = {data_risk1.shape[0]}')
print(f'we lost {100*(removes/data.shape[0])}% of data')
```

```python
Age = data[['age']].mean() #à¸à¸³à¸«à¸™à¸”à¹ƒà¸«à¹‰ Age à¹€à¸›à¹‡à¸™à¸„à¹ˆà¸²à¹€à¸‰à¸¥à¸µà¹ˆà¸¢à¸‚à¸­à¸‡à¸­à¸²à¸¢à¸¸
data[['age']] = data[['age']].fillna(Age) #à¹à¸—à¸™à¸„à¹ˆà¸²à¹€à¸‰à¸¥à¸µà¹ˆà¸¢à¸¥à¸‡à¹ƒà¸™à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸—à¸µà¹ˆà¸§à¹ˆà¸²à¸‡ à¸«à¸£à¸·à¸­à¹€à¸›à¹‡à¸™à¸„à¹ˆà¸² missing
```

```python
data[['risk']] = data[['risk']].fillna('à¹„à¸¡à¹ˆà¹„à¸”à¹‰à¸£à¸°à¸šà¸¸') #à¹à¸—à¸™à¸„à¹ˆà¸² Missing à¸”à¹‰à¸§à¸¢à¸„à¸³à¸§à¹ˆà¸² à¹„à¸¡à¹ˆà¹„à¸”à¹‰à¸£à¸°à¸šà¸¸
```

```python
data[['sex']] = data[['sex']].fillna('à¹„à¸¡à¹ˆà¹„à¸”à¹‰à¸£à¸°à¸šà¸¸') #à¹à¸—à¸™à¸„à¹ˆà¸² Missing à¸”à¹‰à¸§à¸¢à¸„à¸³à¸§à¹ˆà¸² à¹„à¸¡à¹ˆà¹„à¸”à¹‰à¸£à¸°à¸šà¸¸
```

```python
data['risk'].unique() #à¹€à¸¥à¸·à¸­à¸à¸”à¸¹à¸Šà¸·à¹ˆà¸­risk
```

```python
risk_1 = data[data['risk']=='Sandbox'] #à¹€à¸¥à¸·à¸­à¸ Sandbox
risk_1
```

```python
risk_2 = data[data['risk']=='State Quarantine'] #à¹€à¸¥à¸·à¸­à¸ State Quarantine
risk_2
```

```python
risk_3 = data[data['risk']=='Samui plus'] #à¹€à¸¥à¸·à¸­à¸à¹€à¸­à¸²à¹€à¸‰à¸žà¸²à¸° Samui plus
risk_3
```

```python
data_U = pd.concat([risk_1,risk_2,risk_3]) #à¸£à¸§à¸¡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¹‚à¸”à¸¢à¹ƒà¸Šà¹‰ .concat
data_U
```

```python
data_FM = data_U.replace('à¸«à¸à¸´à¸‡',1) #à¸à¸³à¸«à¸™à¸”à¹€à¸žà¸¨à¸«à¸à¸´à¸‡à¹€à¸›à¹‡à¸™1
data_FM
```

```python
data_M = data_FM.replace('à¸Šà¸²à¸¢',150) #à¸à¸³à¸«à¸™à¸”à¹€à¸žà¸¨à¸Šà¸²à¸¢à¹€à¸›à¹‡à¸™150
data_M
```

```python
data_N = data_M.replace('à¹„à¸¡à¹ˆà¹„à¸”à¹‰à¸£à¸°à¸šà¸¸',300) #à¸à¸³à¸«à¸™à¸”à¹„à¸¡à¹ˆà¸£à¸°à¸šà¸¸à¹€à¸žà¸¨à¹€à¸›à¹‡à¸™300
data_N
```

```python
data_risk2 = data_N[['age','risk']] #à¹€à¸¥à¸·à¸­à¸à¹€à¸­à¸²à¹€à¸‰à¸žà¸²à¸°à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸—à¸µà¹ˆà¸ˆà¸°à¸—à¸³à¸à¸£à¸²à¸Ÿ
data_risk2
```
